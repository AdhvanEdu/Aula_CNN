## Operação de convolução

### Realiza o produto cartesiano entre filtros e regiões locais de uma determinada entrada

        

![](img/convolution.gif)

```python
from skimage import io
from scipy.signal import convolve2d
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt


img = io.imread("220px-Albert_Einstein_Head.jpg", as_gray=True)

#define o primeiro filtro
filter01 = np.array([[0,  1, 0],
                     [0, -2, 0],
                     [0,  1, 0]])

#define o segundo filtro
filter02 = np.array([[0,  0, 0],
                     [1, -2, 1],
                     [0,  0, 0]])

#define o terceiro filtro
filter03 = np.array([[0,  0, 0],
                     [1, 0, -1],
                     [0,  0, 0]])

# aplica os filtros, usando a operação de convolução
out01 = convolve2d(img, filter01, mode="valid")
out02 = convolve2d(img, filter02, mode="same")
out03 = convolve2d(img, filter03, mode="same")

plt.subplot(111)
plt.title(img.shape)
plt.imshow(img, cmap='gray')
plt.show()

plt.subplot(111)
plt.title(out01.shape)
plt.imshow(out01, cmap='gray')
plt.show()

plt.subplot(111)
plt.title(out02.shape)
plt.imshow(out02, cmap='gray')
plt.show()

plt.subplot(111)
plt.title(out03.shape)
plt.imshow(out03, cmap='gray')
plt.show()
```

## Pooling
### Reduz o tamanho da entrada através da seleção do atributo máximo ou médio de uma dada parte dos dados. Geralmente, pooling de 2x2 com stride=2 é utilizado (tamanho da saída: entrada/2)



```python
#exemplo de aplicação de average pooling

from skimage.measure import block_reduce

images = [img]

# perform 5 poolings
for i in range(5):
  images.append(block_reduce(images[-1], (2, 2), np.min))

# plot them all
for i in range(6):
  plt.subplot(231 + i)
  plt.imshow(images[i], cmap='gray')

plt.tight_layout()
```

```python
### exemplo de aplicação de max pooling

from skimage.measure import block_reduce

images = [img]

# perform 5 poolings
for i in range(5):
  images.append(block_reduce(images[-1], (2, 2), np.mean))

# plot them all
for i in range(6):
  plt.subplot(231 + i)
  plt.imshow(images[i], cmap='gray')

plt.tight_layout()
```
